### CSV file이란?

> CSV(영어: comma-separated values)는 몇 가지 필드를 쉼표(,)로 구분한 텍스트 데이터 및 텍스트 파일이다. 확장자는 .csv이며 MIME 형식은 text/csv이다. comma-separated variables라고도 한다. 오래전부터 스프레드시트나 데이터베이스 소프트웨어에서 많이 쓰였으나 세부적인 구현은 소프트웨어에 따라 다르다. 그것들을 추가한 형태가 2005년 10월 RFC 4180에서 Informational(IESG의 외부에서 결정된 유용한 정보의 제공)로 사양이 문서화됐다.
> 출처 : wikidocs, https://ko.wikipedia.org/wiki/CSV_(%ED%8C%8C%EC%9D%BC_%ED%98%95%EC%8B%9D)

## Filtering해서 DataFrame 만들기

이 부분이 제일 중요하다.

### **_이 기능은 실제로 airbnb에 있는 기능을 구현하는 코드이다._ (✨ 별표 백만개)**

여러분들은 airbnb나 hotels.com 등등 호텔이나 숙박을 예약할 때, 주로 무엇을 보는가?
가격, 리뷰 갯수 많은 순위, 위치 등등을 여러가지를 고려해서 볼 것이다.

이 때 쓰는 방법이 바로 Filtering 기법이다.

Filtering으로 여러분들이 원하는 숙박업소가 나올 수 있도록 기준치를 정해주면, 사이트는 Dataset 중에서 소비자가 원하는 기준에 맞춰서 여러분들이 원하는 data만 Filtering 해서 보여줄 것이다.

이 기능을 여러분들이 직접 구현해볼 것이다.

    1. 250 달러 이상의 집을 filtering
    2. neighborhood 가 Downtown 인 집을 filtering
    3. reviews 가 5 이상인 집을 filtering

![image.png](./img/3.png)

````python
cost_filter = data['price'] > 250.0
neighborhood_filter = data['neighborhood'] == 'Downtown'
reiviews_filter = data['reviews'] >  5

all_filtering = data[cost_filter & neighborhood_filter & reiviews_filter]
print('Start Filtering Job')
print('The number of house : 17 ')
all_filtering
````

## Price Filtering  한 data visualization

앞서서 여러분들은 Data 재조합 및 Filtering을 통해서 여러분들의 기준에 맞는 data들을 select 했을 것이다.
**Data는 재조합 및 수정한다고 하여 끝나는 것이 아니라, 어떻게 분석하고 해석할 것인가가 굉장히 중요하다.**

따라서 이 분석과 해석을 도와주는 기능 중 하나가 Data Visualization 이다.
Data Visualization을 얼마나 잘하느냐에 따라 여러분들은 흙 속에 진주를 찾을 수도 있고, 못 찾을 수도 있을 것이다.

_아래의 출력물은 Filtering 했던 Data들의 Price 히스토그램을 만드는 출력이다. (그림 및 x, y축 이름, title 등등 출력물과 완전히 똑같이 만들어보자.)_

출력 : 

![image.png](./img/4.png)

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

fig = plt.figure()

ax = fig.add_subplot(1,1,1)
plt.hist(all_filtering['price'], bins=20, alpha = 0.3 , color='r')
plt.xlabel('Price')
plt.ylabel('The number of House')
plt.title('Filtering Price Histogram')

plt.show()
````
