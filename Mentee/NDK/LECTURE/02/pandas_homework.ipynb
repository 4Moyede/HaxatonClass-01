### 4. 선택한 Column의 Value의 종류 파악

```python
airb_select = airb.iloc[:,[2,4,5]]
#직접 컬럼명을 입력하기보단, iloc 메소드를 이용하여 column 번호만 입력해도 나오게...

airb_room = airb_select.iloc[:,0]
# airb_select라는 새로운 df가 생생되었으므로 기존에 사용했던 컬럼번호를 사용하면 x
room_type_value = set(airb_room)

# room_type의 value가 문제를 보면 집합의 형태를 띄고 있으므로 앞서 행했던 중복제거하는 메소드를 굳이  사용하지 않아도
# 집합의 성질때문에 알아서 중복제거가 됨...!

airb_neighborhood = airb_select.iloc[:,1]
neighborhood_value = set(airb_neighborhood)

airb_reviews = airb_select.iloc[:,2]
reviews_value = set(airb_reviews)

select_dict = {'room_type' : room_type_value, 'neighborhood' : neighborhood_value, 'reviews': reviews_value}
print(select_dict)
```

### 5.Filtering 해서 Dataframe 만들기

```python
airb_filter = airb[(airb['price'] > 250) & (airb['neighborhood'] == 'Downtown') & (airb['reviews'] > 5)]
print('''Start Filtering Job
The number of house : 
''',len(airb_filter.index) )

#aib_filter를 len 구문 보다 앞서서 정의했어야...했는데...그러지 못해서 구문실행이 되지 않았다...!

airb_filter
```

### 6. Price Filtering한 data visualization

```python

import matplotlib.pyplot as plt

plt.hist(airb_filter['price'],bins = 20,color = '#ffb2b2')
plt.xlabel('Price') # x축 라벨링
plt.ylabel('The Number of House') # y축 라벨링
plt.title("Filtering Price Histogram", fontsize = 18) #히스토그램 제목
plt.show()

```
