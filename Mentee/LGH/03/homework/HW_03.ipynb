```python
import pandas as pd
import numpy as np
pd.read_csv("airbnb.csv")
```

```python
df = pd.read_csv("airbnb.csv")
del_repetition = df.drop_duplicates("accommodates", keep = 'first') #accommodates에서 중복되는 value는 첫번째 값만 남기고 제거
accommodates_ascending = del_repetition.sort_values(by='accommodates', ascending = True)
accommodates_df = accommodates_ascending['accommodates'] #중복 제거 후 accommodates column만 뽑아서 series 만들기
list_from_accommodates_df = accommodates_df.values.tolist() #리스트로 변환
for i in list_from_accommodates_df : 
    print("This is accommodates's value :",i)
```

```python
good_info = df.loc[:,['room_type','neighborhood','reviews','overall_satisfaction','accommodates','bedrooms','price','minstay']]
print("We select only Good information about airbnb dataset : ")
print("1  :  room_type")
print("2  :  neighborhood")
print("3  :  reviews")
print("4  :  overall_satisfaction")
print("5  :  accommodates")
print("6  :  bedrooms")
print("7  :  price")
print("8  :  minstay")
good_info
```

```python

RoomType = df.loc[:,['room_type']] #room type 출력
roomtype_del_repetition = RoomType.drop_duplicates("room_type", keep = 'first') #room type 처음꺼 남기고 중복 제거
dic_RoomType = roomtype_del_repetition.to_dict() #room type 딕셔너리
Neighborhood = df.loc[:,['neighborhood']] #neighborhood 출력
neighborhood_del_repetition = Neighborhood.drop_duplicates("neighborhood", keep = 'first')#neighborhood 처음꺼 남기고 중복 제거
dic_Neighborhood = neighborhood_del_repetition.to_dict() #neighborhood 딕셔너리
print(dic_RoomType, dic_Neighborhood) #출력,,


```

```python
filtering = (df.price > 250)&(df.neighborhood == 'Downtown')&(df.reviews > 5)
df_filtering = df.loc[filtering, :]
df_filtering
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.xlabel('Price')
plt.ylabel('The number of House')
plt.title('Filtering Price Histogram')
ax.hist(df_filtering['price'], bins=20, color='r', alpha=0.3)

```
